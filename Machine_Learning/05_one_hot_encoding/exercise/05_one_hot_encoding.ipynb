```python
import pandas as pd
```

```python
df = pd.read_csv('D:\\Data Science\\Code basics\\py-master\\ML\\5_one_hot_encoding\\Exercise\\carprices.csv')
df
```

```python
df.columns = [x.lower() for x in df.columns]
df
```

```python
df.rename(columns = {'car model':'car_model','sell price($)':'sell_price', 'age(yrs)':'age_yrs'},inplace=True)
df.columns
```

# Using the dummy variables

```python
dummies = pd.get_dummies(df.car_model)
dummies
```

```python
merged = pd.concat([df,dummies], axis = 'columns')
merged.rename(columns={'Audi A5':'Audi_A5','BMW X5':'BMW_X5','Mercedez Benz C class':'Mercedez_Benz_C_class'}, inplace=True)
merged
```

```python
final = merged.drop(['car_model','Mercedez_Benz_C_class'], axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X = final.drop('sell_price', axis='columns')
X
```

```python
y = final.sell_price
y
```

```python
model.fit(X,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(X,y)
```

# Using Sklearn One Hot Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.car_model = le.fit_transform(dfle.car_model)
dfle
```

```python
X = dfle[['car_model','mileage','age_yrs']]
X
```

```python
y = dfle.sell_price.values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('car_model', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,45000,4]])
```

```python
model.predict([[1,0,86000,7]])
```

```python
model.score(X,y)
```
