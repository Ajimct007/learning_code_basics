```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
## Checking for 1st 5 datas at the digits data
digits.data[:5]
```

```python
df = pd.DataFrame(digits.data)
df.head() # we have the 64 integers which assigned to 64 columns and we have to assign them to a target variable
```

```python
digits.target
```

```python
df['target']=digits.target
```

```python
df.head()
```

```python
## Using the train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'],axis='columns'), digits.target,test_size=0.2)
```

```python
len(X_train), len(X_test)
```

```python
### Using the random forest classifier
from sklearn.ensemble import RandomForestClassifier # ensemple used when we use multiple algorithms to predict the outcome
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test) ## here we got 98% of accuracy
```

```python
## Lets tune the model
model = RandomForestClassifier(n_estimators=20) ## mentioning that 20 random trees
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test) ## here the score improved slightly
```

```python
model = RandomForestClassifier(n_estimators=30) ## mentioning that 30 random trees
model.fit(X_train, y_train)
```

```python
model.score(X_test,y_test) ## here the score improved slightly
```

```python
## plotting confusion matrix to see distribution of errors

y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
```

```python
cm
```

```python
## Visualizing confusion matrix
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
