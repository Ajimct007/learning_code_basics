```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('D:\\Data Science\\Code basics\\py-master\\ML\\1_linear_reg\\Exercise\\canada_per_capita_income.csv')
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'],df['per capita income (US$)'], color='red', marker='+')
```

```python
new_df = np.array(df.year).reshape(-1, 1)
new_df
```

```python
income = np.array(df['per capita income (US$)']).reshape(-1, 1)
income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,income)
```

```python
reg.predict([[2020],[2021]])
```
