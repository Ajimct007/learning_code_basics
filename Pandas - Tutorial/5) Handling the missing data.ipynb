```python
'''
Topics covered
    1) fillna to fill missing values using different ways
    2) interpolate to make a guess on missing values using interpolation
    3) dropna to drop rows with missing values
'''
```

```python
import pandas as pd
# df=pd.read_csv("D:\\Data Science\\py-master\\pandas\\5_handling_missing_data_fillna_dropna_interpolate\\weather_data.csv")

#convert the day column into date column
df=pd.read_csv("D:\\Data Science\\py-master\\pandas\\5_handling_missing_data_fillna_dropna_interpolate\\weather_data.csv",
              parse_dates=["day"])

type(df.day[0])#show the data type of the day -- it was converted as the timestamp

df.set_index('day',inplace=True)#Set day as the index

df
```

```python
#to change the N/A value into 0
new_df = df.fillna(0)
new_df
```

```python
#to specify diff value with diff columns using the dictionary
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

```python
#here for the processing convenience we will fill the 0 value with the previous value
new_df= df.fillna(method='ffill')#N/A value will be carry forward with previous day's value
new_df
```

```python
#for filling the next day's value
new_df= df.fillna(method='bfill')#it will fill the next day's value
new_df
```

```python
#When we say that the axis = 'columns', it will fill data horizontally by backward direction
new_df= df.fillna(method='bfill',axis = "columns")
new_df
```

```python
#for filling the previous day's value for only once use 'limit'

new_df= df.fillna(method='ffill', limit = 1)#it will fill the next day's value
new_df
```

```python
#for filling N/A values that fall inbetween  value

new_df = df.interpolate()

'''interpolate() used to fill missing values in a DataFrame or Series by using various interpolation techniques. 
It helps to estimate or guess the missing values based on the available data.

the default interpolate method is "linear" whcih can be changed into other values

'''
new_df
```

```python
#using the interpolate method = 'time' --- it will change the value which is close to the nearby date
new_df = df.interpolate(method='time')
new_df
```

```python
#to drop the N/A value row
new_df = df.dropna()
new_df
```

```python
#To drop the N/A value if entire row has N/A
new_df = df.dropna(how='all')
#it removed the 9th jan data since the entire row has N/A value
new_df
```

```python
#if i have atleat one value keep that row and drop other valued
new_df = df.dropna(thresh=1)
#Since 9th jan does not has any value it dropped it
new_df
```

```python
#if i have atleat twp value keep that row and drop other valued
new_df = df.dropna(thresh=2)
#it dropped the row which doesnot have atleast 2 values
new_df
```
