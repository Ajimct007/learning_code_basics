```python
#Using replace to change the special values - we may find them various places
import pandas as pd
import numpy as np
df=pd.read_csv("D:\\Data Science\\py-master\\pandas\\6_handling_missing_data_replace\\weather_data.csv")
df
```

```python
#replace the -99999 values with NaN

new_df = df.replace(-99999,np.NaN)
new_df
```

```python
#If we have two special values - we can supply a list
new_df = df.replace([-99999,-88888],np.NaN)
new_df
```

```python
#To replace 0 but we should not do it for the entire column with single go since windspeed can have the 0 value if there is no wind
new_df = df.replace({
    "temperature":-99999,
    "windspeed":-99999,
    "event":"0"
},np.NaN)
new_df
```

```python
#To replace the value specific value -- can also called as mapping using the value
new_df=df.replace({
    -99999:np.NaN,
    "No Event":"Sunny"
})
new_df
```

```python
#replacing the unit value or dropping it
#we can use the regex - they are used to detect patterns
new_df=df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
    },'',regex=True)#This will replace the alphabet value with empty space, i.e remove the unit value
new_df
```

```python
#replacing list with another list
df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df
```

```python
#Replace the alphabet score with the number value
new_df = df.replace(['poor', 'average', 'good', 'exceptional'], [1,2,3,4])
new_df
```
