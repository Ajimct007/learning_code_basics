```python
#Concatenate used to join two or more dataframes

import pandas as pd
#created a DF
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
us_weather
```

```python
#concat used for joining 2 or more DF's
df = pd.concat([india_weather, us_weather])
df #it used the index from original DF
```

```python
df = pd.concat([india_weather, us_weather],ignore_index=True)
df #we can get a continuous index
```

```python
#for giving specific keys for the df based on the country
df = pd.concat([india_weather, us_weather], keys=['india','us'])
df
```

```python
#for appending the 2nd DF as columns insted of rows, mention the index for correct match and join accordingly
temperature_df = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
}, index=[0,1,2])
temperature_df
```

```python
windspeed_df = pd.DataFrame({
    "city": ["delhi","mumbai"],
    "windspeed": [7,12],
}, index=[1,0])
windspeed_df
```

```python
#We have data from 2 cities but the columns are different
#to make the new column
df = pd.concat([temperature_df,windspeed_df],axis=1)
#axis=1 --  means append the data as column
df
```
