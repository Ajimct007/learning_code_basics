```python
#Basic Merge Using a Dataframe Column
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1,df2,on='city')
df3
```

```python
#what is we have 2 diff cities in the DF
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando","baltimore"],
    "temperature": [21,14,35,32],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san franscisco"],
    "humidity": [65,68,71],
})
df2
```

```python
df3 = pd.merge(df1,df2,on='city')
df3
#When we perform the merge it will join only cities which are available in both the DF by leaving the remaining
#in simple terms it did a inter-section or inner join
```

```python
df3 = pd.merge(df1,df2,on='city',how="outer")
df3
#by mentioning how=outer it took all the cities for merging
```

```python
df3 = pd.merge(df1,df2,on='city',how='left')
#left and right is decided by the order
df3
```

```python
df3 = pd.merge(df1,df2,on='city',how='right')
#left and right is decided by the order
df3
```

```python
#indicator flag -- shows the source of the dataframe
df3 = pd.merge(df1,df2,on='city',how='outer',indicator=True)
df3
```

```python
#suffixes
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3=pd.merge(df1,df2,on='city')
#it automatically appended the _x and _y for the temperature and humidity, because the humidity and 
#temperature columns are repeated in the both columns
df3
```

```python
#for providing the suffix insted of taking them automatically
df3=pd.merge(df1,df2,on='city',suffixes=('_left','_right'))
df3
```

```python
df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second'))
df3
```
