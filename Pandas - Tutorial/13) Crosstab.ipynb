**It a table with matrix format that dislays the frequency distribution of the variables**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('D:\\Data Science\\Code basics\\py-master\\pandas\\13_crosstab\\survey.xls')
df.head()
```

```python
## creating the contigency table
pd.crosstab(df.Nationality, df.Handedness) ## the 1st variable cor column and 2nd variable for rows
```

```python
pd.crosstab(df.Sex, df.Handedness) ## the 1st variable cor column and 2nd variable for rows
```

```python
## for generating the total supply margins as true
pd.crosstab(df.Sex, df.Handedness, margins=True)
```

```python
## for having multiple variables
pd.crosstab(df.Sex, [df.Handedness, df.Nationality], margins=True) ## For using the multiple columns with levels
```

```python
pd.crosstab([df.Nationality,df.Sex], [df.Handedness], margins=True)
```

```python
## For getting the percentage value, it ignores the magins
pd.crosstab([df.Sex],[df.Handedness], margins=True, normalize='index')
```

```python
## for getting the average Age
pd.crosstab([df.Sex],[df.Handedness], values=df.Age, aggfunc=np.average)
```
