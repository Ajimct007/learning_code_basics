```python
#What is pivot and pivot table
    #Pivot allows you to transform or reshape data
    
    #Ex - for transforming the data by x-axis into y-axis and the vice-versa (or)
    
    #Changing columns into rows and rows into columns
```

```python
import pandas as pd
df= pd.read_csv("D:\\Data Science\\py-master\\pandas\\10_pivot\\weather.csv")
df
```

```python
#to transform
df.pivot(index='date',columns='city')#index is for rows
```

```python
#to transform and only want the humidity value
df.pivot(index='date',columns='city', values='humidity')#index is for rows
```

```python
#changing the dataframe with requirement
df.pivot(index='humidity',columns='city')
```

```python
#Importing new weather data for pivot table
df = pd.read_csv("D:\\Data Science\\py-master\\pandas\\10_pivot\\weather2.csv")
df
```

```python
#for getting average temperature in the DF
df.pivot_table(index="city",columns='date')
#we got the avg temp of the city for the given date
```

```python
#to sum up the values insted of making them average value
df.pivot_table(index="city",columns='date', aggfunc="sum")
```

```python
df.pivot_table(index="city",columns='date', aggfunc="count")
```

```python
df.pivot_table(index="city",columns='date', aggfunc="diff")
```

```python
df.pivot_table(index="city",columns='date', aggfunc="mean")
```

```python
df.pivot_table(index="city",columns='date', margins=True)
#it added all column and row which shows the average value
```

```python
# Grouper in pivot table

df = pd.read_csv("D:\\Data Science\\py-master\\pandas\\10_pivot\\weather3.csv")
df
#Here we have a single city with multiple dates
```

```python
"""
Here the date is in the string format,we have to convert them into date format for making date as column
"""
df['date']=pd.to_datetime(df['date'])
```

```python
type(df['date'])
```

```python
#for getting avg temp on month basis

'''
here freq=M - means we have used the month end refer the pandas grouping documentation for reference
'''
df.pivot_table(index=pd.Grouper(freq='M', key='date'),columns='city')
```
