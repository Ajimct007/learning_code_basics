```python
import pandas as pd
```

```python
## consider that our csv file does not have a date column
## Date Range function allows us to generate the dates
## Date range allows us to skip the week days for the stock price analysis
```

```python
df = pd.read_csv('D:\\Data Science\\Code basics\\py-master\\pandas\\15_ts_date_range\\aapl_no_dates.csv')
df.head()
```

```python
## for adding the date

rng = pd.date_range(start='6/1/2017', end='6/30/2017', freq='B') ## here B means BUsiness days which skips the weekends
rng
```

```python
#setting the choosen date for our dataframe as index
df.set_index(rng, inplace=True)
df.head()
## we can able to see the date added
```

```python
%matplotlib inline
df.Close.plot()
```

```python
##for specifying the date
df['2017-06-01':'2017-06-10']
```

```python
##Getting the average of the closing price
df['2017-06-01':'2017-06-10'].Close.mean()
```

```python
## for having price for weekends as asuming stock work on weekends
df.asfreq('D', method='pad') ##setting the freq as days with method as pad
## here pad will carry forward the previous values for missing one
```
