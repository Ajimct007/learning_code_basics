```python
#Reading the Excel Files
df = pd.read_excel("D:\\Data Science\\py-master\\pandas\\4_read_write_to_excel\\stock_data.xlsx",sheet_name = "Sheet1")
df
```

```python
#To convert the n.a value to a reasonable value
def convert_people_cell(cell):
    if cell =="n.a.":
        return 'sam walton'
    return cell
def convert_eps_cell(cell):
    if cell == 'not available':
        return None
    return cell
df = pd.read_excel("D:\\Data Science\\py-master\\pandas\\4_read_write_to_excel\\stock_data.xlsx",sheet_name = "Sheet1",
                  converters = {
                      'people':convert_people_cell,
                      'eps':convert_eps_cell
                  })
df
```

```python
#Writing the Excel file
df.to_excel("D:\\Data Science\\py-master\\pandas\\4_read_write_to_excel\\new_aji.xlsx",sheet_name="stocks",startrow=1,startcol=2,index=False)
```

```python
#Writing 2 dataframes in a same excel file with different sheets

df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})

```

```python
with pd.ExcelWriter('D:\\Data Science\\py-master\\pandas\\4_read_write_to_excel\\stocks_weather_aji.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```
