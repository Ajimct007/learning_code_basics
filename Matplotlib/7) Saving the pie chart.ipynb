```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Draw pie chart to track down home expenses
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("D:\\Data Science\\py-master\\matpltlib\\7_save_chart\\piechart_aji.jpg", 
            bbox_inches="tight", pad_inches=1, transparent=True)
'''
bbox_inches -- for providing a box around the chart
pad_inches -- for providing allowance around the chart for decent looking
transparent -- 
    if True -background will be transparent


Note - Refer the documentation
'''
```
