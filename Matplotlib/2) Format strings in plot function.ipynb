```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, 'g+')# Here g means Green and + means the marker for the data point
```

```python
plt.xlabel('Day')
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, 'g+-.')
```

```python
plt.xlabel('Day')
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, 'rD')#for diamond points
```

```python
plt.xlabel('Day')
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y, color="red",marker='D', linestyle='')
```

```python
plt.plot(x,y,color='#888840',marker='+',linestyle='')#color can be given as the hexadecimal code like #888840
```

```python
#for increasing the size of the marker
plt.plot(x,y,color='green',marker='+',linestyle='',markersize=20)
```

```python
plt.plot(x,y,'g<',alpha=1) # alpha can be specified on a scale 0 to 1
#alpha used for the transpancy of the data line or pointer
```

```python
plt.plot(x,y,'g<',alpha=0.5) # alpha can be specified on a scale 0 to 1
#alpha used for the transpancy of the data line or pointer
```
