```python
'''
Matplotlib
    2d plotting library
    for data visualization
    
Why do we need data visualization?
    for better understanding of the data
    
A picture is worth a thousand words

'''
```

```python
import matplotlib.pyplot as plt

#when using Jupyter notebook use the following
%matplotlib inline
```

```python
#The following is the weather data which have the week days with the temperature
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
#to plot a simple 2d plot
plt.plot(x,y)
```

```python
#For changing the colour of the line
plt.plot(x,y,color='green',linewidth=5)
```

```python
plt.plot(x,y,color='green',linewidth=5,linestyle='dashdot')
```

```python
plt.plot(x,y,color='green',linewidth=5,linestyle='dotted')
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y,color='green',linewidth=5,linestyle='dotted')
```

```python
plt.plot(x,y,color='green',linewidth=5,linestyle='--')
```
