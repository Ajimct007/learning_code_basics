```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Weather data
days=[1,2,3,4,5,6,7] #week days
max_t=[50,51,52,48,47,49,46] #max temp
min_t=[43,42,40,44,33,35,37] #min temp
avg_t=[45,48,48,46,40,42,41] #Avg temp
```

```python
#For plotting multiple lines
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
```

```python
# set axes labels and chart title
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

#setting label for providing the legend

plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
```

```python
#providing the legend property
# Show legend
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='best')#to change the location of the legend
#by default the legend set as best which means where we have the empty space, there will be it placed
```

```python
#providing the legend property
# Show legend
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='upper right')#to change the location of the legend
```

```python
#providing the legend property
# Show legend
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='best',shadow=True)#to give a shadow for the legend
```

```python
# Legend at different location with shadow enabled and fontsize set to large
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='best',shadow=True,fontsize='large')
```

```python
#for having grid
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='upper right', fontsize="large",shadow=True)
plt.grid()
```
