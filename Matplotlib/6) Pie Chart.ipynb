```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Draw pie chart to track down home expenses
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
#For getting the good circle 
plt.pie(exp_vals,labels=exp_labels)
plt.axis("equal")
plt.show()
```

```python
#Show percentages for every pie. Specify radius to increase chart size
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.2f%%',radius=1)#in the autopct specify how many decimal points 
#has to be displayed --- if autopct='%0.2%' will display 2 decimal points
#for shadows set Shadow as "True"
plt.show()
```

```python
#Cut the pie chart like the pizza the function is called as "Explode"
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1,explode=[0,0.2,0.2,0.1,0.2])
plt.show()

#Note the pie chart start with the counter clock wise direction
```

```python
#for starting in some angle
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1,explode=[0,0.2,0.2,0.1,0.2],startangle=45)
plt.show()
```
