```python
#Here we have two values for a company
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
#we want to plot the profit on the same chart
plt.xticks(ypos,company)
plt.ylabel("revenue(bln)")
plt.title("US tech stocks")

#Here we have created 2 y values for our bar chart

plt.bar(xpos,revenue,label="Revenue" )
plt.bar(xpos,profit,label='profit')
plt.legend()
#it plots the profit and revenue on a same bar but we have to get the bar on side for better comparision
```

```python
#we want to plot the profit on the same chart
plt.xticks(ypos,company)
plt.ylabel("revenue(bln)")
plt.title("US tech stocks")

#since the x position is same for the both revenue and profit we have to modify them

plt.bar(xpos-0.2,revenue,label="Revenue" )
plt.bar(xpos+0.2,profit,label='profit')
plt.legend()
#it plots the profit and revenue with litter better but we can do more better
```

```python
#we want to plot the profit on the same chart
plt.xticks(ypos,company)
plt.ylabel("revenue(bln)")
plt.title("US tech stocks")

#since the x position is same for the both revenue and profit we have to modify them
#we have to control the width of the bar
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue" )
plt.bar(xpos+0.2,profit,width=0.4,label='profit')
plt.legend()
```

```python
#for Horizontal bar chart

plt.yticks(ypos,company)
plt.title("US tech stocks")

plt.barh(xpos-0.2,revenue,label="Revenue" )
plt.barh(xpos+0.2,profit,label='profit')
plt.legend()
```
